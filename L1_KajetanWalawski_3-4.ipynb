**Zadanie 3** - Zdefiniowanie funkcji *f(x) = x^2 + 5* i sporządzenie jej wykresu dla argumentów z trzech różnych przedziałów.

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def fun(x):
    return x**2 + 5

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)
```

```python
plt.plot(x1, fun(x1), 'b', label='x należy do przedziału (-1,1)')
plt.title("Wykres f(x) = x^2 + 5")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.legend(loc=0)
plt.grid()
plt.show()

plt.plot(x2, fun(x2), 'r', label='x należy do przedziału (-6,6)')
plt.title("Wykres f(x) = x^2 + 5")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.legend(loc=0)
plt.grid()
plt.show()

plt.plot(x3, fun(x3), 'g', label='x należy do przedziału (0,5)')
plt.title("Wykres f(x) = x^2 + 5")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.legend(loc=0)
plt.grid()
plt.show()
```
